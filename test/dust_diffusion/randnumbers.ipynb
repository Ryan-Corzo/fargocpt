```python
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
fig, axes = plt.subplots(ncols=2)

for f in os.listdir("rand"):
    filename = "rand/" + f
    norm, uniform = np.genfromtxt(filename, unpack=True)
    
    ax = axes[0]
    hist, xi = np.histogram(norm)
    x = 0.5*(xi[1:] + xi[:-1])
    ax.plot(x, hist, label=f"{f}")
    
    ax = axes[1]
    hist, xi = np.histogram(uniform)
    x = 0.5*(xi[1:] + xi[:-1])
    ax.plot(x, hist, label=f"{f}")
    
```
